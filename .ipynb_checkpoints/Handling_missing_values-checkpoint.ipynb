```python
import pandas as pd
df = pd.read_csv("Book2.csv", parse_dates =['day'])
df.set_index('day', inplace=True)
df
```

```python
# df.fillna(0)
df.fillna({
    'temp': 0,
    'windspeed' : 0,
    'event': 'No event'
})
```

```python
df.fillna(method="ffill" ,limit=1)
# df.fillna(method="bfill")
```

```python
new_df= df.interpolate(method="time")
new_df
```

```python
new_df= df.dropna()
new_df
```

```python
new_df= df.dropna(how="all")
new_df
```

```python
new_df= df.dropna(thresh=1)
new_df
```
