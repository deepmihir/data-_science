```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
day = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,46,40,42,41,43]
```

```python
plt.xlabel("Days")
plt.ylabel("Tempreture")
plt.title("Weather")
plt.plot(day, max_t, label="Max", marker="o")
plt.plot(day, min_t, label="Min", marker="o")
plt.plot(day, avg_t, label="Avg", marker="o")
plt.legend()
plt.grid()
```
